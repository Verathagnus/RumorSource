So what it does is, we know the infected nodes as given in infected list 
and we return the possible infection source of every infected node in a list of the infectionsource list using the algorithms
Hence non-infected would be adjacent list without containing any of the infected elements.  

````python
'''
The class rumorFinder consists of all the methods required 
to first do the following:-
a) csvReader: function for reading the source destination
              values from a csv file and making an adjacency
              list from it where each list contains the adjacent nodes
              for  each node.

b) selectSource: This is the code to select a pseudo-random
                  node from the other nodes as infection source




c) rumorSpreading: This is the code for using the SI infection model for
                   simulating the spread on infection on the infection sources
                   selected by the selectSource function. The infected nodes
                   are then returned as a list.


d) findRumourCentremot: This is the code for detecting the rumor source using the
                     method of types as given in the relative leaf-ounting algorithm


e) plotGraph: This is used to make the error graph for the errors in estimation of
              rumor source along with increase in number of nodes being infected.

'''

class rumorFinder:     
    def csvReader(self, filename):
        with open(filename, newline='') as f:
            reader = csv.reader(f)
            data = list(reader)
        data=data[1:]
        set_ = {1}
        for line in data:
            set_.add(line[0])
            set_.add(line[1])
        numNodes = len(set_)-1
        adjacency = [[] for _ in range(0, numNodes)]
        for line in data:          
            x, y = int(line[0])-1, int(line[1])-1
            adjacency[x].append(y)
            adjacency[y].append(x)
        return adjacency
    

    def selectSource(self,adjacency):
        src = random.randint(0,len(adjacency)-1)
        while len(adjacency[src]) == 0:
            src = random.randint(0,len(adjacency)-1)
        return src

    def rumorSpreading(self,src,adjacency,numNodesToInfect):
        infectionSource = [[] for _ in range(0,numNodesToInfect)]
        infected = [-1]*numNodesToInfect
        infected[0]=src

        i = 1
        while i < numNodesToInfect:
            susceptible = list()
            srcIdx = list()
            for j in range(0,i):
                neighbour = adjacency[infected[j]]
                for k in range(0,len(neighbour)):
                    if neighbour[k] not in infected:
                        susceptible.append(neighbour[k])
                        srcIdx.append(j)
            randIdx = random.randint(0,len(susceptible)-1)
            infectedNew = susceptible[randIdx]
            src = srcIdx[randIdx]
            infected[i]=infectedNew
            infectionSource[src].append(i)
            infectionSource[i]=[src]
            i += 1
        return infectionSource,infected

    # 2 Helper functions for implementing rumor source finder using ML Source Estimator
    def upwardMsgs(self, upmsg, infectionSource, parent, current):
        if len(infectionSource[current]) == 1:
            upmsg[parent][0] += 1
            upmsg[parent][1] *= upmsg[current][1]
        elif parent == current:
            for i in range(0, len(infectionSource[current])):
                upmsg = self.upwardMsgs(upmsg, infectionSource, current, infectionSource[current][i])
        else:
            for i in range(0, len(infectionSource[current])):
                childNode = infectionSource[current][i]
                if parent != childNode:
                    upmsg = self.upwardMsgs(upmsg, infectionSource, current, childNode)
                    upmsg[current][1] *= upmsg[childNode][1]
            upmsg[current][1] *= upmsg[current][0]
            upmsg[parent][0] += upmsg[current][0]
        return upmsg

    def downwardMsgs(self, msgDown, msgUp, infectionSource, parent, current):
        N = len(infectionSource)
        if current != parent:
            msgDown[current] = msgDown[parent] + log(msgUp[current][0]) - log(N - msgUp[current][0])
            for i in range(0,len(infectionSource[current])):
                if infectionSource[current][i] != parent:
                    msgDown = self.downwardMsgs(msgDown, msgUp, infectionSource, current, infectionSource[current][i])
        else:
            msgDown[current] = log(factorial(N)) - log(N)
            for i in range(0,len(infectionSource[current])):
                msgDown[current] -= log(msgUp[infectionSource[current][i]][1])
            for j in range(0,len(infectionSource[current])):
                msgDown = self.downwardMsgs(msgDown, msgUp, infectionSource, current, infectionSource[current][j])
        return msgDown

    def findRumourCentre(self, infectionSource):
        root, rumorCenter = 0, -1
        N = len(infectionSource)
        upMsgs = [[1,1] for _ in range(0, N)]
        downMsgs = [1] * N                                                                                               
        upMsgs = self.upwardMsgs(upMsgs, infectionSource, root, root)
        downMsgs = self.downwardMsgs(downMsgs, upMsgs, infectionSource, root, root)
        rumorCenter = downMsgs.index(max(downMsgs))
        return rumorCenter



    # Helper functions for finding the infected neighbours of a node from the 
    # adjacency and infected lists
    def uninfectedNeighbours(self, v, adjacency, infected):
        return adjacency[v]-infected[v]
    
    # Helper functions for finding the number of neighbours of a node
    def numNeighbours(self, v):
      return len(adjacency[v])

    def findRumourCentremot(self, infectionSource, infected, adjacency, numNodes):
        root = {}
        leaves = {}
        #Sets is used as uniion
        Nu = {{} for _ in range(0, numNodes)}#Possible error
        for v in infectionSource:
          v.children = {}
          v.armLens = {}
          v.avgLen = {}
        for v in infectionSource:
          Nu[v] = uninfectedNeighbours(v, infected, adjacency)#TODO correction
          v.children = Nu(v)
          v.armLens = armlens(Nu[v])#TODO
          if numNeighbors(v) - len(v.children) == 0:#TODO numNeighbours
            root = v
          if numNeighbors(v) - len(v.children) == 1:
            leaves.union(v)
            v.avgLen = 1
        #Discover Parent, Average Arm Lengths, Update Leaves List
        while leaves != {}:
          v = leaves[1]
          par = Neighbours(v).difference(v.children)#TODO
          (par.children).union(v)
          (par.armLens).union(v.avgLen)
          leaves.difference(v)
          if numNeighbors(par) - len(par.children) == 1:
            par.avgLen = 1+avg(par.armLens)#TODO
            leaves.union(par)
          if numNeighbors(par) - len(par.children) == 0:
            root = par
            leaves.difference(par)
        while max(root.armLens) - 1 > avg((root.armLens).difference(max(root.armLens))):
          maxIdx = index(max(root.armLens))#TODO
          v_cap = root.children[maxIdx]
          (root.children).difference(v_cap)
          (root.armLens).difference(root.armLens[maxIdx])
          root.avgLen = 1 + avg(root.armLens)
          v_cap.children.union(root)
          v_cap.armLens.union(root.avgLen)
          v_cap.avgLen = 1 + avg(v_cap.armLens)
          root = v_cap
          rumorCentre = root
        return rumorCenter
    # Function for plotting the graph
    def plotGraph(self, NMin, NMax, rumorError, NValues, name):
        plt.plot(NValues, [i*100 for i in rumorError], lw = 2.5, label = "Error in Detection")
        plt.axis([NMin, NMax, min(rumorError), 100])
        plt.xlabel("Number of Nodes Infected")
        plt.ylabel("Error in Detecting Source")
        plt.legend(loc = "lower right")
        plt.title("Error in Source Detection")
        plt.grid(which = "major")
        plt.savefig("./csv/"+str(name))

````

````python
'''
This is the actual running of the code for the dolphin.csv
'''
def main():
    rfObj = rumorFinder()
    minNodes = 1
    maxNodes = 20
    xAxis = list()
    rumorRate = list()
    maxIterations = 10
    nameGraph = 'dolphin.png'
    adjacency =  rfObj.csvReader("./csv/dolphin.csv")

    for i in tqdm(range(minNodes, maxNodes+1, 1)):
        numOferrors = 0
        counter, j = 0, 0
        
        while j < maxIterations:
            while 1:
                try:
                    src = rfObj.selectSource(adjacency)
                    infectionSource, infected = rfObj.rumorSpreading(src, adjacency, i)
                    break
                except:
                    continue
            rumorCenter = infected[rfObj.findRumourCentre(infectionSource)]
            #rumorCenter2 = infected[rfObj.findRumourCentre2(infectionSource, infected, adjancency, )]
            if src != rumorCenter:
                numOferrors += 1
            counter += 1
            j += 1
        
        if counter != 0:
            rumorRate.append(numOferrors/counter)
            xAxis.append(i)
    rfObj.plotGraph(minNodes, maxNodes, rumorRate, xAxis, nameGraph)

if __name__ == "__main__":
    main()
````
